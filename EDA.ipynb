### Check the missing values use difference ways

```python
# one way to find a missing vlaues

data_v1.isnull().sum()
```

```python
# Another way to find missing values

for features in data_v1.columns:
    missing_count = data_v1[features].isnull().sum()

    if missing_count > 0:
        print(f'{features}: {missing_count}')
```

```python
# Although this method can use to find a missing values
plt.figure(figsize = (13,6))
sns.heatmap(
    data_v1.isnull(),
    yticklabels = False,
    cbar = False,
    cmap = 'viridis'
)
```

```python
# Calculate missing values per column
missing_values = data_v1.isnull().sum()

# Plot the missing values
plt.figure(figsize=(8, 7))
missing_values[missing_values > 0].plot(kind='bar', color='#1124F2')
plt.title('Missing Values per Column', fontsize=16)
plt.xlabel('Columns', fontsize=14)
plt.ylabel('Number of Missing Values', fontsize=14)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the count plot
plt.figure(figsize=(12, 6))  # Adjust the figure size for better readability
ax = sns.countplot(
    y='Country',             # Use 'y' for horizontal bars, 'x' for vertical bars
    data=data_v3,             # Specify the DataFrame
    order=data_v3['Country'].value_counts().index,  # Order by frequency
    hue = 'Country'
)

# Add labels and title
plt.title('Count of Countries', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Country', fontsize=14)

# Annotate the count on top of each bar
for p in ax.patches:
    width = p.get_width()  # Get the width of the bar (which represents the count)
    plt.text(
        width + 0.1,       # x-position: slightly to the right of the bar
        p.get_y() + p.get_height() / 2,  # y-position: middle of the bar
        f'{int(width)}',   # Text to display (the count)
        ha='left',         # Horizontal alignment
        va='center'        # Vertical alignment
    )

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the top 5 countries
top_5_countries = data_v3['Country'].value_counts().nlargest(3)

# Create labels and sizes for the pie chart
labels = top_5_countries.index  # Country names
sizes = top_5_countries.values  # Counts
explode = (0.1, 0, 0)  # "Explode" the first slice for emphasis

# Create the pie chart
plt.figure(figsize=(10, 10))  # Set the size of the chart
patches, texts, autotexts = plt.pie(
    sizes,                   # Sizes of the pie slices
    labels=labels,           # Labels for each slice
    autopct='%1.2f%%',       # Display percentages with 1 decimal place
    startangle=45,           # Rotate the chart for better readability
    explode=explode,         # Explode the first slice
    shadow=True,             # Add a shadow effect
    colors=plt.cm.Pastel1.colors,  # Use a pastel color scheme
    textprops={'fontsize': 12}  # Set font size for labels and percentages
)

# Adjust the position of percentage values and labels
for text in texts + autotexts:
    text.set_fontsize(12)  # Set font size for better readability

# Add a title
plt.title('Top 5 Countries by Count', fontsize=16, pad=20)

# Add a legend
plt.legend(
    title='Country',         # Legend title
    loc='upper right',       # Position of the legend
    bbox_to_anchor=(1.3, 1), # Adjust legend position outside the chart
    fontsize=12              # Set font size for legend
)

# Show the chart
plt.tight_layout()  # Ensure everything fits in the figure
plt.show()
```

```python
# Check the rating 

ratings = data_v3.groupby(by = ['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating Count'})
```

```python
ratings
```

# Observation
1. When Rating is between 4.5 to 4.9 > Excellent
2. When Rating are between 4.0 to 3.4-->very good
3. when Rating is between 3.5 to 3.9-> good
4. when Rating is between 3.0 to 3.4--> average
5. when Rating is between 2.5 to 2.9 > average
6. when Rating is between 2.0 to 2.4 > Poor

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define a color mapping for the 'Rating color' column
color_mapping = {
    'White': 'black',
    'Red': 'red',
    'Orange': 'orange',
    'Yellow': 'yellow',
    'Green': 'green',
    'Dark Green': 'darkgreen'
}

# Map the 'Rating color' column to valid color codes
ratings['Color Code'] = ratings['Rating color'].map(color_mapping)

# Create the bar plot
plt.figure(figsize=(17, 8))
ax = sns.barplot(
    x='Aggregate rating',  # X-axis: Aggregate rating
    y='Rating Count',      # Y-axis: Rating Count
    data=ratings,          # Data source
    hue='Rating color',    # Hue: Rating color
    palette=color_mapping, # Use the color mapping dictionary
    dodge=False            # Ensure bars are not dodged (grouped)
)

# Add count annotations on top of each bar
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',  # Text to display (count)
        (p.get_x() + p.get_width() / 2., p.get_height()),  # Position of the annotation
        ha='center',               # Horizontal alignment
        va='bottom',               # Vertical alignment
        fontsize=12,               # Font size
        color='black',             # Text color
        xytext=(0, 5),             # Offset for text position
        textcoords='offset points'
    )

# Customize the plot
plt.title('Aggregate Rating Count with Rating Colors', fontsize=16)
plt.xlabel('Aggregate Rating', fontsize=14)
plt.ylabel('Rating Count', fontsize=14)
plt.legend(title='Rating Color', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside

# Show the plot
plt.tight_layout()
plt.show()
```

## Observation 
1. Not Rated count is very high
2. Maximum number of rating rate between 2.6 to 4.3

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the count plot
plt.figure(figsize=(10, 6))
ax = sns.countplot(
    y='Rating color',  # Use 'y' for horizontal bars, 'x' for vertical bars
    data=ratings,      # Data source
    order=ratings['Rating color'].value_counts().index,  # Order by frequency
    hue='Rating color',  # Assign 'Rating color' to hue
    palette=color_mapping,  # Use the color mapping dictionary
    legend=False  # Disable the legend (since y-axis already shows the colors)
)

# Add count annotations on top of each bar
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_width())}',  # Text to display (count)
        (p.get_width() + 0.1, p.get_y() + p.get_height() / 2),  # Position of the annotation
        ha='left',               # Horizontal alignment
        va='center',             # Vertical alignment
        fontsize=12,             # Font size
        color='black'            # Text color
    )

# Customize the plot
plt.title('Count of Rating Colors', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Rating Color', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming your data is stored in a DataFrame called `data_v3`
# Count the number of occurrences of each city
city_distribution = data_v3['City'].value_counts()

# Get the top 4 cities
top_4_cities = city_distribution.head(5)

# Set a professional color palette using seaborn
colors = sns.color_palette('pastel')[0:5]  # Use a pastel color palette

# Create a pie chart with a professional look
plt.figure(figsize=(10, 8))  # Set the size of the pie chart

# Plot the pie chart
wedges, texts, autotexts = plt.pie(
    top_4_cities,
    autopct='%1.1f%%',
    startangle=45,
    colors=colors,
    shadow=True,
    labels=top_4_cities.index,
    explode=(0.05, 0.05, 0.05, 0.05, 0.05),  # Slightly separate the slices
    textprops={'fontsize': 12, 'fontweight': 'bold'}  # Customize percentage text
)

# Add a legend outside the graph
plt.legend(
    wedges,
    top_4_cities.index,
    title="Cities",
    loc="upper left",  # Position of the legend
    bbox_to_anchor=(1, 1),  # Move the legend outside the graph (right side)
    fontsize=12
)

# Add a title
plt.title('Top 4 Cities Distribution', fontsize=16, fontweight='bold', pad=20)

# Add a white circle in the center to make it look like a donut chart (optional)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that the pie is drawn as a circle
plt.axis('equal')

# Display the pie chart
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
